```python
#from models.stacked_lstm_v2 import LSTM_stacked
from models.stacked_rnn import Stacked_RNN

import torch
from torchinfo import summary
import torch.nn as nn

# --- Parâmetros ---
INPUT_SIZE = 10         # Número de features em cada passo de tempo
HIDDEN_SIZE = [64,32]        # Número de unidades em cada camada LSTM
NUM_LAYERS = 2          # <--- Aqui definimos que será uma STACKED LSTM de 2 camadas
OUTPUT_SIZE = 1         # Queremos prever 1 valor (ex: preço futuro)
CELL_TYPE = "gru"

BATCH_SIZE = 32
SEQ_LEN = 5             # Comprimento da sequência de entrada (self.n_past)
```

```python
# --- 1. Instanciação ---
model = Stacked_RNN(
    input_size=INPUT_SIZE, 
    hidden_sizes=HIDDEN_SIZE, 
    output_size=OUTPUT_SIZE,
    cell_type=CELL_TYPE
)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Modelo criado com {NUM_LAYERS} camadas LSTM.")
summary(model, input_size=(BATCH_SIZE, SEQ_LEN, INPUT_SIZE))
```

```python
# --- 2. Criação de Dados de Exemplo ---
# Dados de entrada no formato: (Batch, Seq_Len, Input_Size)
dummy_input = torch.randn(BATCH_SIZE, SEQ_LEN, INPUT_SIZE)
dummy_input = dummy_input.to(device)
display(dummy_input.shape)
```

```python
# --- 3. Execução ---
prediction = model(dummy_input)

print("\n--- Resultados ---")
print(f"Formato da entrada: {dummy_input.shape}")
print(f"Formato da previsão: {prediction.shape}")
print(f"A previsão é para {OUTPUT_SIZE} valor(es) para cada item do Batch.")
```
